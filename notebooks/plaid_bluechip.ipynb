```python
def get_stock(symbol: str):
    response = ddb_table.query(
        KeyConditionExpression=Key('pk').eq('STOCK#{}'.format(symbol)) &  Key('sk').begins_with('IND#') 
    )
    return response
```

```python
def create_account(email, item) -> dict:
    '''Transform item to put into DDB'''
    dt = datetime.utcnow()
    item['pk'] = 'USER#{}'.format(email)
    item['sk'] = 'ACC#{}'.format(item['account_id'])
    item['date_added'] = dt.strftime("%m/%d/%Y")
    r = ddb_table.put_item(
        Item=item
    )
    return {'Account Created': item['sk'], 'Response': r}
email = "test@gmail.com"
item = {
"account_id":"123",
"name":"testAcc",
"official_name": "Official Name",
"type":"Invest",
"subtype":"IRA",
"holdings":"",
"investment_transactions" :""
}
item['pk'] = 'USER#{}'.format(email)
item['sk'] = 'ACC#{}'.format(item['account_id'])
r = ddb_table.put_item(Item=item)
```
